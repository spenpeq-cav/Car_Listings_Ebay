# Cleaning Columns

```python
autos.columns
```

```python
autos.columns = ['date_crawled', 'name', 'seller', 'offer_type', 'price', 'ab_test',
       'vehicle_type', 'registration_year', 'gearbox', 'power_ps', 'model',
       'odometer', 'registration_month', 'fuel_type', 'brand',
       'unrepaired_damage', 'ad_created', 'num_photos', 'postal_code',
       'last_seen']
```

```python
autos.columns
```

To make the columns easier to read and access, we have changed the names to the more preferred snakecase

```python
autos.describe(include='all')
```

After investigating the columns, the 'price' and 'odometer' columns should be numberic values but is stored as text that needs cleaning. The 'price' column contains '$' and ',' characters that need to be removed. The 'odometer' column contains 'km' and a ',' character that need removing. The name will be changed to 'odometer_km' for more clarity.

```python
autos['price'] = autos['price'].str.replace('$','').str.replace(',','').astype(int)
autos['odometer'] = autos['odometer'].str.replace('km','').str.replace(',','').astype(int)
autos.rename({'odometer': 'odometer_km'}, axis= 1, inplace= True)
```

# Dealing with Incorrect Registation Year Data

The registration of cars should fall into the years 1900 - 2016. Anything outside this interval could be removed. We will count the number of listings outside of this interval to see if it's safe to remove the rows entirely.

```python
(~autos["registration_year"].between(1900,2016)).sum() / autos.shape[0]
```

Given that only about 4% of our data is outside of our interval of 1900 - 2016, it is safe to remove.

```python
autos = autos[autos["registration_year"].between(1900,2016)]
autos["registration_year"].value_counts(normalize=True).head(10)
```

Most of the vehicles were registered within the past 20 years.

# Exploring Price by Brand

```python
brand_percentages = autos["brand"].value_counts(normalize=True) * 100
brand_percentages
```

Volkswagen is by far the most common brand. We will limit further analysis to only brands with a significant percentage.

```python
common_brands = brand_percentages[brand_percentages > 5].index
common_brands
```

```python
brand_mean_prices = {}

for brand in common_brands:
    brand_only = autos[autos["brand"] == brand]
    mean_price = brand_only["price"].mean()
    brand_mean_prices[brand] = int(mean_price)

brand_mean_prices
```

There is a significant gap between the top brands. Audi, BMW, Mercedes Benz are the most expensive whereas ford and opel are the cheapest. Volkswagen is more middle of the pack and can be seen as a 'best of both worlds' kind of car.

# Exploring Mileage

We will calculate the mean mileage and construct a Dataframe containing the mean price and mileage for the top brands.

```python
brand_mean_mileage = {}

for brand in common_brands:
    brand_only = autos[autos["brand"] == brand]
    mean_miles = brand_only["odometer_km"].mean()
    brand_mean_mileage[brand] = int(mean_miles)

brand_mean_mileage
```

```python
mean_mileage = pd.Series(brand_mean_mileage).sort_values(ascending=False)
mean_price = pd.Series(brand_mean_prices).sort_values(ascending=False)
```

```python
brand_info = pd.DataFrame(mean_mileage, columns=['mean_mileage'])
```

```python
brand_info
```

```python
brand_info['mean_price'] = mean_price
brand_info
```

The mean mileage range does not differ as much as the mean prices do. There is a slight trend with the more expensive cars having more mileage.
